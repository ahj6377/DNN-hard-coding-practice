```python
import numpy as np
import tensorflow as tf
import os
import matplotlib.pyplot as plt


from tensorflow.keras.layers import Dense, Flatten, MaxPool2D,Conv2D
from tensorflow.keras import layers, Sequential, Model
from tensorflow.keras.utils import to_categorical
```

```python
fashion_mnist = tf.keras.datasets.fashion_mnist
```

```python
(train_X,train_y) , (test_X,test_y) = fashion_mnist.load_data()


train_X = train_X/255
test_X = test_X/255

train_y = to_categorical(train_y)
test_y = to_categorical(test_y)

train_X = np.expand_dims(train_X,axis = -1)
test_X = np.expand_dims(test_X,axis=-1)
```

```python
num_classes = 10
input_dim = train_X[0].shape

print(input_dim)
```

```python
model = Sequential()

model.add(Conv2D(filters = 16, kernel_size = (3,3),padding = "same", input_shape = input_dim, activation = "relu"))
model.add(MaxPool2D(pool_size = (2,2), strides = (2,2)))

model.add(Conv2D(filters = 32, kernel_size = (3,3),padding = "same",activation = "relu"))
model.add(MaxPool2D(pool_size = (2,2),strides = (2,2)))

model.add(Conv2D(filters = 64,kernel_size = (3,3), padding = "same", activation = "relu"))
model.add(Conv2D(filters = 64,kernel_size = (3,3), padding = "same", activation = "relu"))
#model.add(MaxPool2D(pool_size=(2,2),strides = (2,2)))        minst의 픽셀 사이즈가 작기에 해당 레이어를 삭제

model.add(Conv2D(filters = 128, kernel_size = (3,3),padding = "same",activation = "relu"))
model.add(Conv2D(filters = 128, kernel_size = (3,3),padding = "same",activation = "relu"))
#model.add(MaxPool2D(pool_size=(2,2),strides = (2,2)))        minst의 픽셀 사이즈가 작기에 해당레이어를 삭제

model.add(Conv2D(filters = 128, kernel_size = (3,3),padding = "same",activation = "relu"))
model.add(Conv2D(filters = 128, kernel_size = (3,3),padding = "same",activation = "relu"))
model.add(MaxPool2D(pool_size=(2,2),strides = (2,2)))

model.add(Flatten())
model.add(Dense(32,activation = "relu"))
model.add(Dense(32,activation = "relu"))
model.add(Dense(num_classes,activation = "softmax"))
```
