```python
def shift(x,sr):
  random_factor = 5
  octaves = 1 + random.random() *random_factor
  
  
  
  out = librosa.effects.pitch_shift(x,sr,octaves)
  
  return out
```

```python
fullseq = np.array([])
for i in range(len(letters)):
  if(letters[i] != ' '):
    let = gTTS(letters[i],lang = lang)
    let.save('./samples/%s.mp3'%letters[i])

  
    path = './samples/%s.mp3'%letters[i]
    x,sr = librosa.load(path)
    newx = newx[np.abs(newx) > 0.00001]
    print(newx)
  else:
    x = np.zeros(16000)
  
  newsound = shift(x,sr)
  
  
  fullseq = np.append(fullseq,newsound)
  
```
